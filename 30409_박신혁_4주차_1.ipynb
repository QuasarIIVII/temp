[실습3] 수학, 영어, 과학 점수의 합계와 평균을 구한 값으로 total과 average 변수(파생변수)를 정수형으로 만들고, 변수가 잘 만들어졌는지 결과를 출력하시오.  
f["변수명"]= 수식              
※ 수식에서 /는 실수형 나누기,  //는 정수형 나누기        

```python
 # [실습3] 아래에 코드를 완성하고 실행하시오.
f["total"]=f["math"]+f["english"]+f["science"]
f["average"]=f["total"]/3
f[["total","average"]]
```

```python
 # [실습4] 아래에 코드를 완성하고 실행하시오.
f["average"].describe()
```

[실습5] 학생의 average 변수를 이용하여 반 전체 평균을 구하여 변수에 저장하고 출력하시오.

```python
# [실습5] 아래에 코드를 완성하고 실행하시오.
total_mean=f["average"].mean()
total_mean
```

[실습6] 세 과목의 평균값이 전체 평균보다 높은 경우 "pass", 아니면 "fail"을 저장하는 "result" 변수를 생성하고 결과를 확인하시오.  
※ numpy.where(조건, 참일때, 거짓일때)   
  
import numpy as np

f.head()  

```python
# [실습6] 아래에 코드를 완성하고 실행하시오.
import numpy as np
f["result"]=np.where(f["average"]>=total_mean, "pass", "fail")      
f.head()  
```
